```python
# All imports
import pandas as pd
import pickle
import numpy as np
```

### READ IN data file:

```python
# FileName = 'tiny.csv'
FileName = 'aaaPas 2024-05-15.csv'
# Read in Data File
df = pd.read_csv(FileName)
display(df.head(4))
```

### Cutting and preprocessing:

```python
# Cut by number of patients
num_of_patients = 500
df_cut = df.loc[df['val_lopnr'] <= num_of_patients]
print('Selected dataframe has a shape of \t {}'.format(df_cut.shape))
```

### WRITE OUT data file:

```python
save_when_1 = 0
save_when_1 = 1
# filename_str = "foo"
filename_str = "aaaPas_first_{}".format(num_of_patients)

save_pickle_filename = filename_str + ".pickle"
save_csv_filename =  filename_str + ".csv"

if save_when_1:
    with open(save_pickle_filename, 'wb') as fh:
        pickle.dump(df_cut, fh)
    df.to_csv(save_csv_filename, index=False)
```
